# <b><u> Problem 1: </u> Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor. </b>

```python
# Your code here
imdb_df.columns
```

```python
columns_to_keep = ['budget','genres','imdb_id','original_title','cast','crew','revenue']
```

```python
imdb_df1 = imdb_df[columns_to_keep]
```

```python
# changing string to a list(columns = cast,crew,genres)
def string_to_list(str):
  return eval(str)
```

```python
imdb_df1.loc[~imdb_df1['cast'].isna(),'cast'] = imdb_df1.loc[~imdb_df1['cast'].isna(),'cast'].apply(string_to_list)
```

```python
imdb_df1.loc[~imdb_df1['crew'].isna(),'crew'] = imdb_df1.loc[~imdb_df1['crew'].isna(),'crew'].apply(string_to_list)
```

```python
imdb_df1.loc[~imdb_df1['genres'].isna(),'genres'] = imdb_df1.loc[~imdb_df1['genres'].isna(),'genres'].apply(string_to_list)
```

```python
imdb_df1.describe()
```

```python
# replace low budget value with median 
imdb_df1.loc[imdb_df1['budget'] < 1000, 'budget'] = imdb_df1.loc[imdb_df1['budget'] >= 1000 ,'budget'].median()
imdb_df1.loc[imdb_df1['revenue'] < 1000, 'revenue'] = imdb_df1.loc[imdb_df1['revenue'] >= 1000 ,'revenue'].median()
```

```python
imdb_df1['profit'] = imdb_df1['revenue'] - imdb_df1['budget']
```

```python
imdb_not_null_df = imdb_df1.dropna()
```

```python

actor_list = []
for i,j in imdb_not_null_df.iterrows():
  for ele in (imdb_not_null_df.loc[i,'cast']):
    if type(ele) == dict:
      actor_list.append(ele['name'])
```

```python
top_actor_name = pd.DataFrame(actor_list,columns = ['actors_name']).value_counts().reset_index().head(3)
top_actor_name
```

```python
no_null_df = imdb_df1.dropna()
```

```python
movie1 = []
movie2 = []

profit1 = []
profit2 = []

for index,j in no_null_df.iterrows():
  for ele in (no_null_df.loc[index,'cast']):
    if type(ele) == dict:
      actor = ele['name']
      if 'Samuel L. Jackson' in actor:
        movie1.append(no_null_df.loc[index,'original_title'])
        profit1.append(no_null_df.loc[index,'profit'])

      if 'Robert De Niro' in actor:
        movie2.append(no_null_df.loc[index,'original_title'])
        profit2.append(no_null_df.loc[index,'profit'])  
```

```python
#creating a loop to get the genres for Robert and Samuel
gener_s=[]
a=[]
for i in range(len(movie1)):
  for g in no_null_df.loc[i,'genres']:
    a.append(g['name'])
    
  gener_s.append(a)
  a=[]

gener_r=[]
b=[]
for i in range(len(movie2)):
  for g in no_null_df.loc[i,'genres']:
    b.append(g['name'])
    
  gener_r.append(b)
  b=[]
```

```python
# creating dataframe for Samuel L. Jackson in genre,movie,profit

mov1 = pd.DataFrame(movie1,columns = ['movie_name'])
prof1 = pd.DataFrame(profit1,columns = ['total_profit'])
gen1 = pd.DataFrame(np.array(gener_s).flatten(),columns = ['genre'])
```

```python
movies_by_samul_jackson = pd.concat([mov1,prof1,gen1],axis =1)
movies_by_samul_jackson.sort_values('total_profit',ascending = False)
```

```python
# creating dataframe for robert d niro in genre,movie,profit

mov2 = pd.DataFrame(movie2,columns = ['movie_name'])
prof2 = pd.DataFrame(profit2,columns = ['total_profit'])
gen2 = pd.DataFrame(np.array(gener_r).flatten(),columns = ['genre'])
```

```python
Robert_De_Niro = pd.concat([mov2,prof2,gen2],axis =1)
Robert_De_Niro.sort_values('total_profit',ascending = False)
```

# <b><u> Problem 2 :</u> Top 3 directors prefer which actors the most?

```python
# Your code here 
def actor_name(index):
  for ele in no_null_df.loc[index,'cast']:
    if type(ele) == dict :
      return ele['name']
```

```python
def director_name(index):
   for ele in no_null_df.loc[index,'crew']:
       if ele['job'] == 'Director':
         
         return ele['name']
```

```python
table_director = pd.DataFrame(columns = ['Director_name','actor_name'])
```

```python
for index,j in no_null_df.iterrows():
  try:
    table_director = table_director.append({'Director_name' : director_name(index),'actor_name' : actor_name(index)},ignore_index = True )
  except:
    continue  
```

```python
table_director
```

```python
table_director['Director_name'].value_counts().rename_axis('directors').reset_index(name = 'counts').head(3)
```

```python
top_3_direc = table_director[table_director['Director_name'].isin(['Ron Howard','Clint Eastwood','Steven Spielberg'])]
top_3_direc
```

```python
actors_list_ron = []
actors_list_clint = []
actors_list_steven = []
for index,row in top_3_direc.iterrows():
  if top_3_direc.loc[index,'Director_name'] == 'Ron Howard':
    actors_list_ron.append(top_3_direc.loc[index,'actor_name'])
  if top_3_direc.loc[index,'Director_name'] == 'Clint Eastwood':
    actors_list_clint.append(top_3_direc.loc[index,'actor_name'])
  if top_3_direc.loc[index,'Director_name'] == 'Steven Spielberg':
    actors_list_steven.append(top_3_direc.loc[index,'actor_name'])
```

```python
actor_prefer_ron_howard = list(set(actors_list_ron))
actor_Clint_Eastwood = list(set(actors_list_clint))
actor_prefer_Steven_Spielberg = list(set(actors_list_steven))
```
